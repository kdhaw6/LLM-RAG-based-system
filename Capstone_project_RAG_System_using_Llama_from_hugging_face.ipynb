This project is a Streamlit-based chatbot application designed to provide AI based tech support through conversational AI. The chatbot uses a language model to generate responses based on user input and converts the responses to speech using the Text-to-Speech API.

# Features

1.   Conversational AI: Provides responses based on user input.
2.   Text-to-Speech: Converts the chatbot's responses into spoken words.
1.   Document Retrieval: Loads and processes PDF documents for context-aware responses.









# RAG

"At its core, this application implements a Retrieval-Augmented Generation, or RAG, system. What does that mean? Imagine you have a set of documents – in our case, they are PDF files containing technical information. Instead of just relying on a language model's general knowledge, which might be outdated or incomplete for specific domains, we first retrieve relevant snippets from these documents based on your question. Then, we feed both your question and these relevant snippets to the language model, which then generates a more informed and contextually accurate answer. This approach combines the power of information retrieval with the generative capabilities of large language models
